**Convert multiple pdf tables to multiple csv file**

```python
path = "/home/jayanthikishore/Desktop/Analysis/Work/ML_EIT/Data/pdf2tables/"

outpath = "/home/jayanthikishore/Downloads/"

all_files = glob.glob(os.path.join(path, "dat*.pdf"))
# display(all_files)

ii=0
for pdffle in all_files[0:]:
    ii+=1
    lngth = len(pdffle)
    csvfle = outpath + pdffle[lngth-9:lngth-4]+'.csv'
    display(pdffle,csvfle)
    tabula.convert_into(pdffle,csvfle,pages="all")
    
```
